```python
#pkg = mp.jupyter.open_package()
pkg = mp.jupyter.open_source_package()
pkg
```

```python
vl = pkg.reference('vl_source').dataframe()
vl
```

```python
m = { r.code:r.label for inx, r in vl[vl.variable == 'race_ethnicity'].iterrows() }
num = { r.code:r.label for inx, r in vl[vl.variable == 'race_ethnicity'].iterrows() if r.code.isnumeric()}
alpha = { r.code:r.label for inx, r in vl[vl.variable == 'race_ethnicity'].iterrows() if r.code.isalpha()}

na = {}

# Check that the number of alpha codes refer to the same thing
for n, a in list(zip(num, alpha)):
    assert(m[a] == m[n])
    
    na[n] = a

def clean(refname):
    
    df = pkg.reference(refname).dataframe()
    df['Race/Ethnicity'] = df['Race/Ethnicity'].replace(na)

    return df    
    
homicides = clean('homicides_s')
homicide_justifiable = clean('homicide_justifiable_s')
homicide_manslaughter = clean('homicide_manslaughter_s')
```
